# Prompt template

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import HumanMessage, SystemMessage, BaseMessage
from langchain_openai import ChatOpenAI

chat_model = ChatOpenAI(model="gpt-4o-mini")

SYSTEM_MESSAGE_CONTENT = """\
# Identity
Bạn là chuyên gia OpenAI am hiểu việc tạo prompt template với kiến thức sâu rộng APE.

# Instructions
1. Giúp tôi tạo prompt template <content> trong thẻ xml <user_prompt>
2. Nếu có gì không rõ ràng hãy hỏi lại tôi, cho đến khi bạn có đầy đủ nội dung để tạo dữ liệu.\
"""

chain = chat_model | StrOutputParser()
messages: list[BaseMessage] = [
    SystemMessage(content=SYSTEM_MESSAGE_CONTENT),
]

for _ in range(20):
    content = input("Enter your content:")
    if content == "n":
        break
    messages.append(HumanMessage(content=content))
    response = chain.invoke(messages)
    print(response)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI
import asyncio

instruction = input("Enter an instruction")
chat_model = ChatOpenAI(model="gpt-4o-mini")

times = int(input("Enter số lần bạn muốn chạy"))

HUMAN_MESSAGE_CONTENT = f"""\
1. Read the following instruction and confirm whether its purpose is understandable:
  > {instruction.encode("unicode_escape").decode()}
2. Trình bày cách bạn hiểu instruction trên. Có ví dụ minh hoạ rõ ràng.
"""

chain = chat_model | StrOutputParser()
coroutines = [
    chain.ainvoke([HumanMessage(content=HUMAN_MESSAGE_CONTENT)]) for _ in range(times)
]

responses = await asyncio.gather(*coroutines, return_exceptions=True)
for response in responses:
    print("*" * 80)
    print(response)
```
